```python
import json
import random
import sys
import time
from pprint import pprint

import requests

sys.path.append("..")
from bohour.poem_samples_large import samples


random.seed(1)

qasaed = [random.choice(samples) for _ in range(10)]
```

```python
for qaseeda in qasaed:
    pprint(qaseeda)
    payload = ""
    for bait in qaseeda:
        payload += "\n".join(bait.split("#"))
        payload += "\n"
    print("*" * 80)
    print(payload)
    print("*" * 80)
    response = requests.post(
        "http://34.123.169.118:8000/api/analyze",
        data={"baits": payload},
    )
    time.sleep(5)
    if response.ok:
        pprint(response.json())
```
